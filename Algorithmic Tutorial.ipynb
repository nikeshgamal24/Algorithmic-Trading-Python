```python
ret = close.pct_change().dropna()
ret.head()
```

```python
ret.describe()
```

```python
ret.describe().T
```

```python
summary = ret.describe().T.loc[:,["mean","std"]]
summary
```

```python
summary["mean"]  = summary["mean"]*252
summary["std"] = summary["std"]*np.sqrt(252)
summary
```

```python
summary.plot.scatter(x="std",y="mean", figsize=(10,5),fontsize=12)
for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,"std"]+0.02,summary.loc[i,"mean"]+0.02),size=12) #plot the values

plt.xlabel("Annual Risk (std)",fontsize=12)
plt.ylabel("Annual Return ",fontsize=12)
plt.title("Return/Risk",fontsize=12)
plt.show()
```
